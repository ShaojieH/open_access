```python
from get_usnews_ranking_2019 import data_path
import json
from fuzzywuzzy import fuzz, process
```

```python
def get_rank_from_name(name, data):
    # names = [d['name'] for d in data]
    # return name, process.extractOne(name, names)
    
    max_ratio = 0
    max_index = -1
    for index, row in enumerate(data):
        if row['name'] in name:
            return index, 200, name, data[index]['name']
        ratio = fuzz.ratio(row['name'], name)
        if ratio > max_ratio:
            max_ratio = ratio
            max_index = index
    if max_ratio >= 40:
        print (max_index, max_ratio, name, data[max_index]['name'])
    
```

```python
us_ranking_path = data_path + "ranking/usnews_ranking_2019_result.json"
ieee_country_path = data_path + "ieee/usa_data_2018.json"
with open(us_ranking_path) as f:
    rank_data = json.load(f)
with open(ieee_country_path) as f:
    ieee_data = json.load(f)
```

```python
for row in ieee_data:
    get_rank_from_name(row['affiliation_name'].lower(), rank_data)
```
